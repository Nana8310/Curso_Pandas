Ejercicio 10. Aplicación de combine_first para completar datos de prueba con datos de entrenamiento:

```python
import pandas as pd
import numpy as np
```

Datos

```python
# Datos de entrenamiento
train_data = {
    'Producto': ['A', 'B', 'C', 'D'],
    'Ventas': [100, 150, 200, 250]
}

# Datos de prueba con valores faltantes
test_data = {
    'Producto': ['A', 'B', 'C', 'E'],
    'Ventas': [np.nan, 160, np.nan, 180]
}
```

Pregunta 10.1: Crear los dataframes

```python
df_train = pd.DataFrame(train_data)
df_test = pd.DataFrame(test_data)
print("Datos de entrenamiento:")
print(df_train)
print("\nDatos de prueba:")
print(df_test)
```

Pregunta 10.2: Combine los datos y guarde los resultados como df_test_combined

```python
# Combinar los DataFrames usando combine_first
df_test_combined = df_train.set_index('Producto').combine_first(df_test.set_index('Producto')).reset_index()
print("Datos combinados:")
print(df_test_combined)
```

1. df_train.set_index('Producto')
Esta parte del código establece la columna ‘Producto’ como el índice del DataFrame df_train. Esto es útil para combinar los DataFrames basándose en esta columna.

2. .combine_first(df_test.set_index('Producto'))
Aquí, combine_first se usa para combinar df_train con df_test, que también tiene ‘Producto’ como índice. La función combine_first toma los valores de df_train y, donde haya valores faltantes (NaN), los rellena con los valores correspondientes de df_test.

3. .reset_index()
Finalmente, reset_index se usa para restablecer el índice del DataFrame combinado, devolviendo ‘Producto’ a ser una columna normal en lugar de un índice.

Resumen del proceso
Establecer índices: Ambos DataFrames (df_train y df_test) tienen ‘Producto’ como índice.
Combinar: combine_first combina los DataFrames, rellenando los valores faltantes en df_train con los valores de df_test.
Restablecer índice: reset_index convierte el índice ‘Producto’ de nuevo en una columna.
El resultado es un DataFrame df_combined que contiene los datos combinados de df_train y df_test, con los valores faltantes rellenados donde sea posible.
